# Merge

```python
import pandas as pd
```

Read in dataset from UFC website

```python
ufc_scrape = pd.read_csv("ufc_website.csv") # read in ufc website scrapes

ufc_scrape.head(5)
```

Next the dataframe from Wikipedia was read in.

```python
fighters = pd.read_csv("wiki_scrape.csv")
fighters.tail(10)
```

```python
fighters['Name'] = fighters['Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
fighters.head(5)
```

The dataframes from the wikipedia page and the UFC website were merged.

```python
complete_wiki_plus_ufc =  (pd.merge(fighters, ufc_scrape, on='Name', how = 'left'))
complete_wiki_plus_ufc = complete_wiki_plus_ufc.dropna(thresh = 13)
complete_wiki_plus_ufc.head(5)
```

We found a csv with the rankings for each weightclass. 

```python
rank = pd.read_csv("rankings.csv")
rank['Rank'] = pd.to_numeric(rank['Rank'], downcast='integer')
rank.head(5)
```

The rankings datset was merged with the datset above. We merged it on name and to the left as not all the fighters are ranked and in the rankings csv

```python
main_with_ranks =  (pd.merge(complete_wiki_plus_ufc, rank, on='Name', how = 'left')) 
main_with_ranks.head(5)
```

The ex_champ data (from the second wikipedia scrape), was read in

```python
ex_champ = pd.read_csv("ex_champ.csv")
```

The ex_champ dataset was merged with the main dataset

```python
main_with_ex_champs =  (pd.merge(main_with_ranks, ex_champ, on='Name', how = 'left'))
main_with_ex_champs.head(5)
```

The merged dataframe was exported as a csv file

Fill Nan values with false

```python
main_with_ex_champs['ex_champ'] = main_with_ex_champs['ex_champ'].fillna(False)
```

```python
main_with_ex_champs.to_csv('merged.csv',index = False)
```
