```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import html5lib

import requests 
from bs4 import BeautifulSoup
import urllib
import time
```

The UFC website has a page for each fighter on the roster, with many statistics that we wish to obtain for this project. The websites urls made up of the base 'https://www.ufc.com/athlete/' followed by a fighters name, seperated by a '-'. For example 'https://www.ufc.com/athlete/conor-mcgregor'. 

So to scrape the website I first needed a list of every fighters name. I found this on the Wikipedia page below:

```python
a = pd.read_html('https://en.wikipedia.org/wiki/List_of_current_UFC_fighters', flavor = 'html5lib')
```

Each weight class was seperated into seperate tables so I put each table into a datframe and merged them. 

```python
a11 = a[11]
a12 = a[12]
a13 = a[13]
a14 = a[14]
a15 = a[15]
a16 = a[16]
a17 = a[17]
a18 = a[18]
a19 = a[19]
a20 = a[20]
a21 = a[21]
a22 = a[22]
 
fighters = pd.concat([a11,a12,a13,a14,a15,a16,a17,a18,a19,a20,a21,a22]) # I called the dataframe fighters
```

The column headers where in the first row so this had to be changed

```python
fighters.columns = fighters.iloc[0]
fighters = fighters.drop(fighters.index[[0]]) 
fighters.reset_index(drop = True)
```

The names needed to be cleaned with certain useless characters stripped.

```python
fighters['Name'] = fighters['Name'].map(lambda x: x.split('*')[0].strip())
fighters['Name'] = fighters['Name'].map(lambda x: x.split('(C)')[0].strip()) # champs had c's beside their names
fighters['Name'] = fighters['Name'].map(lambda x: x.split('(IC)')[0].strip()) # interim champs had ICs

fighters.head(5)
```

Next we had to split the names of each fighter into seperate list. First name in first_names , second names in second_names etc

```python
first_names = []
second_names = []
third_names = []
fourth_names = []


dfToList = fighters['Name'].tolist()

names=[x.split() for x in dfToList]
for name in names:
    first_names.append(str.lower(name[0]))
    if len(name)==2:
        second_names.append(str.lower(name[1]))
    elif len(name)==3:
        second_names.append(str.lower(name[1]))
        third_names.append(str.lower(name[1]))
    elif len(name)==4:
        second_names.append(str.lower(name[1]))
        third_names.append(str.lower(name[1]))
        fourth_names.append(str.lower(name[1]))
```

Its now time to scrape the website. First of all I created two dictionarys. Reach had to be put in a seperate dictionary because the physical attributes were broken on the website with some fighters stats missing. Reach could then have been in either one of two columns. We decided to scrape both and drop the usless data.

All variables were initilised being equal to 'NONE'. Before this if a fighter had one stat missing, it would copy that stat from the previous fighter

The scraper was put in a try,except statement as if one stat was missing before this, the scraper would crash. Now it carries on with the stat left as 'None'.

```python
dfObj = pd.DataFrame(columns=['names', 'k_wins', 'sb_wins','d_wins','hd_strk','lg_stk', 'bd_strk', 'st_land', 'st_attemp'
                                 , 'tk_land', 'tk_attemp','strksL_perm','strksAb_perm','takedAv_per','subAvg_perm',
                                 'strk_d','take_d','kd_rat','av_time' ,'stand_stk','clinch_stk','grnd_stk'])

dfObj2 = pd.DataFrame(columns=[ 'names','reach' , 'reach2']) # declaring both dictionaries


base = 'https://www.ufc.com/athlete/' # base url for every page
for name in names :
    page = base + '-'.join(name) # every name needed to be seperated by a dash
    page1 = requests.get(page)
    
    time.sleep(5)  # pause for 5 seconds so no IP ban
    c = page1.content 
    soup = BeautifulSoup(c,"lxml")
  
    try:
        
            head_strk=None    
            body_strk=None
            ko_wins = None
            sub_wins = None
            dec_wins = None
            head_strk = None
            leg_strk = None
            strks_land = None
            strks_attem = None
            take_land = None
            take_attem = None 
            strksLan_permin = None
            strksAbs_permin = None
            takedAvg_permin = None
            subAvg_permin = None 
            strk_def = None 
            take_def = None
            kd_ratio = None 
            av_time = None
            Reach=None
            Reach2 = None
            
            
            
            name = soup.find(class_ ="field field-name-name").get_text().strip() #get the text and strip white space 

          


            try: # try , except statement

                head_strk = soup.find(id="e-stat-body_x5F__x5F_head_value").get_text().strip() #main body of scraper, looks more complicated than it is
                body_strk = soup.find(id="e-stat-body_x5F__x5F_body_value").get_text().strip()
                leg_strk = soup.find(id = "e-stat-body_x5F__x5F_leg_value").get_text().strip()


                numb1 = soup.findAll(class_ = "c-overlap__stats-value")   # class = tags from html on website

                strks_land = numb1[0].get_text().strip() # multiple values at same tag, require findALL and arrays
                strks_attem = numb1[1].get_text().strip()


                numb3 = soup.findAll(class_ = "c-stat-compare__number")

                strksLan_permin = numb3[0].get_text().strip()
                strksAbs_permin = numb3[1].get_text().strip()
                takedAvg_permin = numb3[2].get_text().strip()
                subAvg_permin = numb3[3].get_text().strip()
                strk_def1 = numb3[4].get_text().strip()
                (strk_def1.split())
                strk_def = strk_def1[0] + strk_def1[1]

                take_def1 = numb3[5].get_text().strip()
                (take_def1.split())
                take_def = take_def1[0] + take_def1[1]

                kd_ratio = numb3[6].get_text().strip()
                av_tim = numb3[7].get_text().strip()




                numb4 = soup.findAll(class_ = "c-stat-3bar__value")

                stand = numb4[0].get_text().strip()
                clinch = numb4[1].get_text().strip()
                ground = numb4[2].get_text().strip()

                ko_wins1  = numb4[3].get_text().strip()
                (ko_wins1.split())
                ko_wins = ko_wins1[0] + ko_wins1[1]
                ko_wins = int(ko_wins)

                sub_wins1 = numb4[5].get_text().strip()
                (sub_wins1.split())
                sub_wins = sub_wins1[0] + sub_wins1[1]
                sub_wins = int(sub_wins)

                dec_wins1 = numb4[4].get_text().strip()
                (dec_wins1.split())
                dec_wins = dec_wins1[0] + dec_wins1[1]
                dec_wins = int(dec_wins)

                numb5 = soup.findAll(class_ = "e-chart-circle__percent")
                take_land =  numb1[2].get_text().strip()
                take_attem = numb1[3].get_text().strip()
                
                numb1 = soup.findAll(class_ = "c-bio__text")  

                Reach = numb1[6].get_text().strip()
                Reach2 = numb1[7].get_text().strip()
      
    
            except: 
                    print('exception')
            





        # Append rows in Empty Dataframe by adding dictionaries
            dfObj = dfObj.append({'names': name, 'k_wins': ko_wins, 'sb_wins': sub_wins,'d_wins':dec_wins,'hd_strk': head_strk, 
                                  'bd_strk' : body_strk, 'lg_stk':leg_strk,'st_land':strks_land, 'st_attemp':strks_attem,
                              'tk_land':take_land, 'tk_attemp' : take_attem, 'strksL_perm': strksLan_permin, 
                              'strksAb_perm': strksAbs_permin , 'takedAv_per' : takedAvg_permin , 'subAvg_perm' :subAvg_permin ,
                          'strk_d' : strk_def , 'take_d': take_def , 'kd_rat': kd_ratio , 'av_time' :av_tim,
                             'stand_stk': stand, 'clinch_stk':clinch, 'grnd_stk': ground}, ignore_index=True)
            
            dfObj2 = dfObj2.append({'names': name, 'reach':Reach , 'reach2': Reach2}, ignore_index=True)

    except Exception as e:
        print(page, " doesn't exist!") # says which pages don't exist

dfObj
```

This dropna staement requires 10 non empty values. Any row with more than this is dropped.

```python
dfObj = dfObj.dropna(thresh = 10)
dfObj.reset_index(drop = True)
dfObj.rename(columns={'names':'Name'}, inplace=True)
dfObj
```
